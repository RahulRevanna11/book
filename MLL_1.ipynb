```python
import pandas as pd
```

```python
df = pd.read_csv('home_dataset.csv')
```

```python
print("Shape of the dataframe:", df.shape)
```

```python
print("\nColumn names:", df.columns)
```

```python
print("\n Display 5 quantiles of the dataset:")
quantiles = df.describe()
print(quantiles)
```

```python
print("\nCount of rows with null values:")
print(df[df.isnull().any(axis=1)].shape[0])
```

```python
print("\nFirst 10 rows:")
print(df.head(10))
```

```python
print("\nLast 10 rows:")
print(df.tail(10))
```

```python
print("\nRows with number of floors >= 1.5:")
print(df[df['floors'] >= 1.5])
```

```python
print("\nHouses older than 1968:")
older_houses = df[df['yr_built'] < 1968]
print(older_houses)
print("\nCount of older houses:", older_houses.shape[0])
```

```python
#Display houses having price more than 50000/- and their count
print("\nHouses with price > 50000:")
expensive_houses = df[df['price'] > 50000]
print(expensive_houses)
print("\nCount of expensive houses:", expensive_houses.shape[0])
```

```python
#Display houses having number of bedrooms less than or equal to 2 and price above 51000/- along with their count
print("Houses having number of bedrooms less than or equal to 2 and price above 51000/- along with their count:")
selected_houses = df[(df['bedrooms'] <= 2) & (df['price'] > 51000)]
print(selected_houses)
print("Count:", selected_houses.shape[0])
print("\n")
```

```python
# Display houses having number of bedrooms more than 2 and square foot
print("\nHouses with number of bedrooms > 2 and square foot:")
selected_houses2 = df[(df['bedrooms'] > 2) & (df['sqft_living'] > 0)]
print(selected_houses2)
```

```python
#Replace/eliminate missing values

df_cleaned = df.dropna()  
print(df_cleaned)
print("\n")
```

```python
df.rename(columns={'sqft_living': 'living_area', 'sqft_lot' : 'lot_area', 'sqft_basement' : 'basement_area'}, inplace=True)
```

```python
df.columns
```

```python
# Drop unessential columns (feature selection)
# For example, dropping 'view' and 'waterfront'
df.drop(['view', 'waterfront'], axis=1, inplace=True)
```

```python
df.columns
```

```python
numeric_columns = df.select_dtypes(include=['number'])

# Calculate the mean of numeric columns
mean_values = numeric_columns.mean()

# Calculate the minimum of numeric columns
min_values = numeric_columns.min()

# Calculate the maximum of numeric columns
max_values = numeric_columns.max()

# Display the results
print("Mean values:\n", mean_values)
print("\nMinimum values:\n", min_values)
print("\nMaximum values:\n", max_values)
```

```python
print("Find mode of all columns:")
print(df.mode())
print("\n")
```

```python
print("Display unique values in each column:")
for column in df.columns:
    print(f"{column}: {df[column].unique()}")
print("\n")
```
