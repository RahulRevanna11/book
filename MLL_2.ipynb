```python
# 8. Perform dimensionality reduction using PCA and t-SNE
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.impute import SimpleImputer

# Select all numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns

# Step 1: Check for Missing Values
missing_values = df[numeric_cols].isnull().sum()
# print("Missing Values:\n", missing_values)

# Step 2: Impute Missing Values
imputer = SimpleImputer(strategy='mean')
df_imputed_values = imputer.fit_transform(df[numeric_cols])


# Create DataFrame with imputed values and original columns
df_imputed = pd.DataFrame(df_imputed_values, columns=numeric_cols)
# print(df_imputed)

pca = PCA(n_components=2)
tsne = TSNE(n_components=2)

df_pca = pd.DataFrame(pca.fit_transform(df_imputed[numeric_cols]), columns=['PC1', 'PC2'])
df_tsne = pd.DataFrame(tsne.fit_transform(df_imputed[numeric_cols]), columns=['TSNE1', 'TSNE2'])

# Compare visual plots
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(df_pca['PC1'], df_pca['PC2'])
plt.title('PCA')

plt.subplot(1, 2, 2)
plt.scatter(df_tsne['TSNE1'], df_tsne['TSNE2'])
plt.title('t-SNE')

plt.show()
```

```python
# 9. Create datasets and plot m vs execution time for PCA and t-SNE
import timeit

def generate_dataset(m, n):
    return np.random.randn(n, m)

def measure_execution_time(func, *args):
    start_time = timeit.default_timer()
    func(*args)
    return timeit.default_timer() - start_time

m_values = [10, 50, 100, 500, 1000, 1500]
n = 10000

pca_execution_times = []
tsne_execution_times = []

for m in m_values:
    dataset = generate_dataset(m, n)

    # Perform PCA
    pca_time = measure_execution_time(PCA().fit, dataset)
    pca_execution_times.append(pca_time)

    # Perform t-SNE
    tsne_time = measure_execution_time(TSNE().fit, dataset)
    tsne_execution_times.append(tsne_time)

    
# Plot m vs execution time
plt.plot(m_values, pca_execution_times, label='PCA')
plt.plot(m_values, tsne_execution_times, label='t-SNE')
plt.xlabel('Number of Features (m)')
plt.ylabel('Execution Time (seconds)')
plt.title('PCA and t-SNE Execution Time Comparison')
plt.legend()
plt.show()
```
